```python
import numpy as np
import pandas as pd
```

```python
ev_charging_reports = pd.read_csv("dataset/ev_charging_reports.csv", delimiter=";")
ev_charging_reports.head()
```

```python
ev_charging_reports["Start_plugin_hour"] = ev_charging_reports["Start_plugin"].str.split(" ").str[0] + " " + ev_charging_reports["Start_plugin_hour"].astype(str) + ":00"
ev_charging_reports.head()
```

```python
print(ev_charging_reports.columns)
```

```python
print(ev_charging_reports.info())
```

```python
ev_charging_reports = pd.get_dummies(ev_charging_reports, columns=["month_plugin", "weekdays_plugin", "User_type"], dtype=float)
print(ev_charging_reports.info())
```

```python
ev_charging_reports.head()
print(ev_charging_reports.shape)
```

```python
traffic_reports = pd.read_csv("dataset/local_traffic_distribution.csv", delimiter=';')
cols_dict = {
    'KROPPAN BRU': 'Kroppan_bru_traffic',
    "MOHOLTLIA": "Moholtlia_traffic",
    "SELSBAKK": "Selsbakk_traffic",
    "MOHOLT RAMPE 2": "Moholt_rampe_2_traffic",
    "Jonsvannsveien vest for Steinanvegen": "Jonsvannsveien_vest_steinanvegen"
}
traffic_reports = traffic_reports.rename(columns=cols_dict)
print(traffic_reports.info())
traffic_reports.head()
```

```python
ev_charging_traffic = ev_charging_reports.merge(traffic_reports, left_on="Start_plugin_hour", right_on="Date_from")
ev_charging_traffic.head()
```

```python
print(ev_charging_traffic.info())
```

```python
drop_columns = ['session_ID', 'Garage_ID', 'User_ID', 
                'Shared_ID',
                'Plugin_category','Duration_category', 
                'Start_plugin', 'Start_plugin_hour', 'End_plugout', 'End_plugout_hour', 
                'Date_from', 'Date_to']

ev_charging_traffic = ev_charging_traffic.drop(columns=drop_columns, axis=1)
ev_charging_traffic.head()
```

```python
print(ev_charging_traffic.info())
```

```python
for column in ev_charging_traffic.columns:
    if ev_charging_traffic[column].dtype == 'object':
        ev_charging_traffic[column] = ev_charging_traffic[column].str.replace(',', '.')
        ev_charging_traffic[column] = ev_charging_traffic[column].str.replace('-', '0')


ev_charging_traffic.head()
```

```python
print(ev_charging_traffic.info())
```

```python
for column in ev_charging_traffic.columns:
    ev_charging_traffic[column] = ev_charging_traffic[column].astype(float)

ev_charging_traffic = ev_charging_traffic.dropna()
ev_charging_traffic.head()
```

```python
numerical_features = ev_charging_traffic.drop(['El_kWh'], axis=1).columns
X = ev_charging_traffic[numerical_features]

y = ev_charging_traffic['El_kWh']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,train_size=0.80, test_size=0.20, random_state=42) # set a random seed - do not modify

print("Training size:", X_train.shape)
print("Testing size:", X_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

linear_model = LinearRegression()
linear_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error

linear_test_predictions = linear_model.predict(X_test)
test_mse = mean_squared_error(y_test, linear_test_predictions)
print("Linear Regression - Test Set MSE:", test_mse)
```

```python
import torch
from torch import nn
from torch import optim


# Convert training set
X_train_tensor = torch.tensor(X_train.values, dtype=torch.float)
y_train_tensor = torch.tensor(y_train.values, dtype=torch.float).view(-1,1)

# Convert testing set
X_test_tensor = torch.tensor(X_test.values, dtype=torch.float)
y_test_tensor = torch.tensor(y_test.values, dtype=torch.float).view(-1,1)
```

```python
torch.manual_seed(42)

# model = nn.Sequential(
#     nn.Linear(27, 56),
#     nn.ReLU(),
#     nn.Linear(56, 27),
#     nn.ReLU(),
#     nn.Linear(27, 1)
# )

model = nn.Sequential(
    nn.Linear(27, 56),
    nn.ReLU(),
    nn.Linear(56, 112),
    nn.ReLU(),
    nn.Linear(112, 56),
    nn.ReLU(),
    nn.Linear(56, 27),
    nn.ReLU(),
    nn.Linear(27, 1)
)
```

```python
loss = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.0007)
```

```python
num_epochs = 10000 # number of training iterations
for epoch in range(num_epochs):
    outputs = model(X_train_tensor) # forward pass 
    mse = loss(outputs, y_train_tensor) # calculate the loss 
    mse.backward() # backward pass
    optimizer.step() # update the weights and biases
    optimizer.zero_grad() # reset the gradients to zero

    # keep track of the loss during training
    if (epoch + 1) % 500 == 0:
        print(f'Epoch [{epoch + 1}/{num_epochs}], MSE Loss: {mse.item()}')
```
